## 1.Cargar el Dataset de Autos y realizar las Estadísticas descriptivas básicas y medidas de tendencia central

```python
autos = pd.read_csv("autos2.csv", encoding="ISO-8859-1")
```

```python
autos.head()
```

```python
autos.info()
```

```python
#Determinar el valor mínimo de precios
autos.price.min()
```

```python
#Determinar el valor máximo de precios
autos.price.max()
```

```python
#Determinar el valor de la moda de precios
moda=autos.price.mode()
moda
```

```python
#Determinar el valor de la media de precios
media=autos.price.mean()
media
```

```python
#Determinar el valor de la mediana de precios
mediana=autos.price.median()
mediana
```

```python
#Determinar el valor de la desviación estándar de precios
desviacion=autos.price.std()
desviacion
```

```python
#Determinar los estadísticos y cuartiles con la función describe()
estcuar=autos.price.describe()
estcuar
```

## 2.Realizar un Histograma de precios de los automóviles


```python
#Determinar mediante la regla de sturges la cantidad optima de bins o separadores
n=len(autos.price)
k=int(1+math.log2(n))
print("Numero de bins:",k )
```

```python
#Poner títulos y encabezados apropiados
plt.hist(autos.price, bins=8)
plt.title('Histograma de Precios de los autos')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

Interprete los datos y proporcione sus conclusiones.

conclusiones:
La mayoría de los autos analizados tienen precios accesiles, aunque algunos son bastante caros, probablemente se trate de modelos de lujo. La distribución no es uniforme, ya que hay una mayor concentración de autos en los rangos de precios más bajos y solo unos pocos en los más altos. Esto sugiere que la mayoría de los autos son accesibles, pero hay excepciones con precios muy elevados.

## 3.Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)

```python
#Poner títulos y encabezados apropiados
sns.boxplot(x='body-style',y='price',data=autos)
```

Interprete los datos y proporcione sus conclusiones

conclusiones:El tipo de vehículo tiene mucho que influir en su precio. Los convertibles y sedanes pueden tener precios bastante altos, mientras que los hatchbacks y wagons suelen ser más económicos. pero aun asi se encuentras exepciones de autos más caros que el promedio. Esto puede deberse a diferencias en la marca o el nivel de lujo.
